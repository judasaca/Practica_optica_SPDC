```python
import matplotlib.pylab as plt
import numpy as np 
import matplotlib.cm as cm
```

```python
archivos=["tipo1filtrohorizontalcolineal.TXT",
"tipo1filtrohorizontal.TXT",
"tipo1filtroverticalcolineal.TXT",
"tipo1filtrovertical.TXT",
"tipo1sinfiltrocolineal.TXT",
"tipo1sinfiltro.TXT",
"tipo2filtrohorizontal.TXT",
"tipo2filtrovertical.TXT",
"tipo2sinfiltro.TXT"]
```

```python
def graficador(data_file_name,png_name,mostrar=False):
    datos= np.loadtxt(data_file_name)
    #limpieza de datos
    promedio=np.mean(datos)
    ancho=80
    limite_inferior=promedio-ancho
    limite_superior=promedio+ancho
    for i in range(len(datos)):
        for k in range(len(datos[0])):
            if datos[i][k] >=limite_superior:
                datos[i][k]=limite_superior
            elif datos[i][k] <=limite_inferior:
                datos[i][k]=limite_inferior
    
    
    #generar grafica
    fig, ax = plt.subplots(figsize=(18,10))
    cax = ax.imshow(datos, cmap=cm.inferno)
    cbar = fig.colorbar(cax)
    ax.set_xlabel("Pixel X",fontsize=20)
    ax.set_ylabel("Pixel Y",fontsize=20)
    plt.xticks(fontsize = 20)
    plt.yticks(fontsize = 20)
    cbar.ax.tick_params(labelsize=20)
    plt.savefig(png_name)

    if not mostrar:
        plt.close()
    return datos
```

```python
for file in archivos:
    g=graficador(file,file[:-4]+".png")
```

```python
d1=graficador(archivos[5],"prueba.png",True)
d2=graficador(archivos[7],"prueba2.png",True)
```

```python
print(np.mean(d1),np.mean(d2),np.std(d1),np.std(d2))
```

```python
datos1=[]
datos2=[]
for i in d1:
    u=i.tolist()
    datos1.extend(u)
for i in d2:
    u=i.tolist()
    datos2.extend(u)
```

```python
plt.hist(datos1,bins=20,density=True)
plt.hist(np.mean(datos1),bins=1,density=True)
plt.ylim(0,0.015)
```

```python
plt.hist(datos2,bins=20,density=True)
plt.hist(np.mean(datos2),bins=1,density=True)
plt.ylim(0,0.0125)
```
